```python
from __future__ import with_statement
matrixPath = 'run140/Abasis_test_N28.bin'
def file_byte_iterator(path):
    """given a path, return an iterator over the file
    that lazily loads the file
    """
    with open(path, 'rb') as file:
        for chunk in file:
            for byte in chunk:
                yield byte
```
